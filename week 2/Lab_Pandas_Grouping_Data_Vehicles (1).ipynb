# Pandas: grouping

```python
import pandas as pd
import numpy as np
```

```python
cars = pd.read_csv("data/vehicles.csv")
```

```python
cars.info()
```

How many Car models? 

```python
cars.groupby('Model').count()
#3608 car models
```

group by the data by the Make  using count function

```python
cars.groupby('Make').count()
```

Converting Grams/Mile to Grams/Km

1 Mile = 1.60934 Km

Converting Gallons to Liters

1 Gallon = 3.78541 Liters

What brand has the most cars?

```python
cars.groupby('Make')['Make'].count().sort_values(ascending=False)
#Chevrolet
```

<b>show the average CO2_Emission_Grams/Km  by Brand

```python
cars['CO2 Emission Grams/Mile']=cars['CO2 Emission Grams/Mile']/1.60934
cars = cars.rename(columns={'Make':'Brand','CO2 Emission Grams/Mile':'CO2 Emission Grams/Km'})
```

```python
cars.groupby('Brand')[['CO2 Emission Grams/Km']].mean()
```

<b>show the average CO2_Emission_Grams/Km  by Brand ... sorted

```python
cars.groupby('Brand')[['CO2 Emission Grams/Km']].mean().sort_values(by='CO2 Emission Grams/Km',ascending=False)
```

# (Optional) 

Use `pd.cut` or `pd.qcut` to create 4 groups (bins) of cars, by Year. We want to explore how cars have evolved decade by decade.

```python
cars['century'] = pd.cut(x=cars['Year'], bins=[1980,1990,2000,2010,2020],
                    labels=['1980','1990','2000','2010'])
```

### Did cars consume more gas in the eighties?

show the average City_Km/Liter by year_range

```python
cars['City MPG']=cars['City MPG']*0.42514285
cars = cars.rename(columns={'City MPG':'City_Km/Liter'})
```

```python
cars.groupby('century')[['City_Km/Liter']].mean()
```

Which brands are more environment friendly?

```python
cars
```

```python
cars.groupby('Brand')[['CO2 Emission Grams/Km']].mean().sort_values(by='CO2 Emission Grams/Km')
```

```python
#by brands
cars.groupby(['century','Brand'])[['CO2 Emission Grams/Km']].mean().sort_values(by='CO2 Emission Grams/Km')
```

Does the drivetrain affect fuel consumption?

```python
cars['Highway MPG']=cars['Highway MPG']*0.42514285
```

```python
cars = cars.rename(columns={'Highway MPG':'Highway_Km/Liter'})
```

```python
cars.groupby('Drivetrain')[['Highway_Km/Liter','City_Km/Liter']].mean().sort_values(by='City_Km/Liter',ascending=False)
```

Do cars with automatic transmission consume more fuel than cars with manual transmission?

```python
cars["Auto/Manuel"]=np.where(cars['Transmission'].str.contains('Auto'), "AUTO","MANUEL")
```

```python
cars.groupby('Auto/Manuel')[['City_Km/Liter']].mean()
```

Use `groupby` and `agg` with different aggregation measures for different columns:

aggregate with average City_Km/Liter and the count of the Trans

```python
cars.groupby('Auto/Manuel')['City_Km/Liter','Auto/Manuel'].agg(["mean","count"])
```

```python
cars['Transmission1']=cars.Transmission.str.extract('([-+]?\d*\.?\d+)')
```

aggregate with average City_Km/Liter and the minimum of the Trans

```python
cars.groupby(['Auto/Manuel','Transmission1'])['City_Km/Liter'].agg([min,np.mean])
```
